```python
def inverseSum(input):
    i = 0
    sum = 0
    wordString = str(input)
    while i < len(str(wordString))-1:
        if wordString[i] == wordString[i+1]:
            sum = sum + int(wordString[i])
        i = i + 1
    if wordString[0] == wordString[len(str(wordString))-1]:
        sum = sum + int(wordString[0])
    return sum
```

```python
def inverseSum2(input):
    i = 0
    sum = 0
    wordString = str(input)
    distance = int(len(wordString)/2)
    while i < distance:
        if wordString[i] == wordString[i+distance]:
            sum = sum + int(wordString[i])*2
        i = i + 1
    return sum
```
